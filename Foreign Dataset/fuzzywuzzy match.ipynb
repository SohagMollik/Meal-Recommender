```python
import pandas as pd
from fuzzywuzzy import process

# Load the dataset
dataset_path = "F:/4th year 2nd semester/Thesis/Dataset/test_dataset.csv"
df = pd.read_csv(dataset_path)

# Function to find similarity using fuzzywuzzy
def find_similarity(user_input, titles):
    # Use process.extract to get similarity scores and titles
    similarity_scores = process.extract(user_input, titles, limit=len(titles))
    # Sort by similarity scores in descending order
    similarity_scores.sort(key=lambda x: x[1], reverse=True)
    return similarity_scores

# Get user input
user_input = input("Enter a food title: ")

# Find similarity with the "title" column
similarity_scores = find_similarity(user_input, df['title'])

# Display the results
print(f"Top 5 similar titles for '{user_input}':")
for title, score in similarity_scores[:5]:
    # Get the corresponding row from the dataframe
    row = df[df['title'] == title].iloc[0]
    # Display the row
    print(row)
    print("-" * 40)
```

```python
import pandas as pd
from fuzzywuzzy import fuzz, process
from prettytable import PrettyTable

# Load the dataset
dataset_path = "F:/4th year 2nd semester/Thesis/Dataset/test_dataset.csv"
df = pd.read_csv(dataset_path)

# Function to find similarity using fuzzywuzzy token_sort_ratio
def find_similarity(user_input, titles):
    # Use process.extract to get similarity scores, titles, and indices with token_sort_ratio
    similarity_scores = process.extract(user_input, titles, scorer=fuzz.token_sort_ratio, limit=len(titles))
    # Sort by similarity scores in descending order
    similarity_scores.sort(key=lambda x: x[1], reverse=True)
    return similarity_scores

# Get user input
user_input = input("Enter a food title: ")

# Find similarity with the "title" column using token_sort_ratio
similarity_scores = find_similarity(user_input, df['title'])

# Create a PrettyTable for displaying the results
result_table = PrettyTable()
result_table.field_names = ["Title", "Energy", "Fat", "Protein", "Salt", "Saturates", "Sugars", "Similarity (%)"]

# Display the top 200 similar titles
for title, score, index in similarity_scores[:500]:
    # Get the corresponding row from the dataframe using the index
    row = df.iloc[index]
    # Calculate the similarity percentage
    similarity_percentage = score
    # Add a row to the PrettyTable
    result_table.add_row([row['title'], row['energy'], row['fat'], row['protein'], row['salt'], row['saturates'], row['sugars'], similarity_percentage])

# Print the result table
print(result_table)

# Save the results to a CSV file
result_table_df = pd.DataFrame(result_table._rows, columns=result_table.field_names)
result_table_df.to_csv("output_results.csv", index=False)
```
