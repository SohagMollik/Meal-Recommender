```python
import pandas as pd
from keras.models import load_model

# Load the saved model
model = load_model("model_test.h5")

# Load the test dataset
test_dataset_path = r"F:\4th year 2nd semester\Thesis\Dataset\cosine_similarity.csv"
test_data = pd.read_csv(test_dataset_path)
test_data = test_data.drop(['energy'], axis=1)
test_data = test_data.drop(['similarity(%)'], axis=1)

# Extract features for prediction (excluding 'title' column)
X_test = test_data.iloc[:, 1:].values

# Make predictions
predictions = model.predict(X_test)

# Threshold for categorizing as Allowed or Restricted
threshold = 0.5

# Initialize lists for Allowed and Restricted foods
allowed_food = []
restricted_food = []

# Categorize foods based on predictions
for i in range(len(test_data)):
    title = test_data.iloc[i, 0]
    prediction = predictions[i][0]  # Assuming a single output neuron for binary classification

    if prediction >= threshold:
        allowed_food.append(title)
    else:
        restricted_food.append(title)

# Print the results
print("Allowed food:")
for i, food in enumerate(allowed_food, start=1):
    print(f"{i}. {food}")

print("\nRestricted food:")
for i, food in enumerate(restricted_food, start=1):
    print(f"{i}. {food}")
```
