```python
import pandas as pd
from fuzzywuzzy import fuzz, process
from prettytable import PrettyTable

# Load the dataset
dataset_path = r"C:\Users\ASUS\Downloads\BD test data.csv"
df = pd.read_csv(dataset_path)


# Function to find similarity using fuzzywuzzy token_sort_ratio
def find_similarity(user_input, food_names):
    # Use process.extract to get similarity scores, titles, and indices with token_sort_ratio
    similarity_scores = process.extract(user_input, food_names, scorer=fuzz.token_sort_ratio, limit=len(food_names))
    # Sort by similarity scores in descending order
    similarity_scores.sort(key=lambda x: x[1], reverse=True)
    return similarity_scores

# Get user input
user_input = input("Enter a food title: ")

# Find similarity with the "title" column using token_sort_ratio
similarity_scores = find_similarity(user_input, df['Food name in Bengali'])

# Create a PrettyTable for displaying the results
result_table = PrettyTable()
result_table.field_names = ["Food name in Bengali", "Energy", "Protein", "Fat", "Carbohydrate available", "Total dietary fibre", "Ash", "Na", "Similarity (%)"]

# Display the top 200 similar titles
for title, score, index in similarity_scores[:500]:
    # Get the corresponding row from the dataframe using the index
    row = df.iloc[index]
    # Calculate the similarity percentage
    similarity_percentage = score
    # Add a row to the PrettyTable
    result_table.add_row([row['Food name in Bengali'], row['Energy'], row['Protein'], row['Fat'], row['Carbohydrate available'], row['Total dietary fibre'], row['Ash'], row['Na'], similarity_percentage])

# Print the result table
print(result_table)

# Save the results to a CSV file
result_table_df = pd.DataFrame(result_table._rows, columns=result_table.field_names)
result_table_df.to_csv("Fuzzy_match_results.csv", index=False)
```

```python
import pandas as pd
from fuzzywuzzy import fuzz, process
from prettytable import PrettyTable

# Load the dataset
dataset_path = r"C:\Users\ASUS\Downloads\BD test data.csv"
df = pd.read_csv(dataset_path)

# Strip whitespaces from column names and convert them to lowercase
df.columns = df.columns.str.strip().str.lower()

# Function to find similarity using fuzzywuzzy token_sort_ratio
def find_similarity(user_input, food_names):
    similarity_scores = process.extract(user_input, food_names, scorer=fuzz.token_sort_ratio, limit=len(food_names))
    similarity_scores.sort(key=lambda x: x[1], reverse=True)
    return similarity_scores

# Get user input
user_input = input("Enter a food title: ")

# Find similarity with the "title" column using token_sort_ratio
similarity_scores = find_similarity(user_input, df['food name in bengali'])

# Create a PrettyTable for displaying the results
result_table = PrettyTable()
result_table.field_names = ["Food name in Bengali", "Energy", "Protein", "Fat", "Carbohydrate available", "Total dietary fibre", "Ash", "Na", "Similarity (%)"]

# Display the top 200 similar titles
for title, score, index in similarity_scores[:500]:
    row = df.iloc[index]
    similarity_percentage = score
    result_table.add_row([row['food name in bengali'], row['energy'], row['protein'], row['fat'], row['carbohydrate available'], row['total dietary fibre'], row['ash'], row['na'], similarity_percentage])

# Print the result table
print(result_table)

# Save the results to a CSV file
result_table_df = pd.DataFrame(result_table._rows, columns=result_table.field_names)
result_table_df.to_csv("Fuzzy_match_results.csv", index=False)
```
