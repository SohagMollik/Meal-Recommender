```python
import pandas as pd

# Load the dataset
file_path = r"C:\Users\ASUS\Downloads\BD train data.csv"
dataset = pd.read_csv(file_path)

# Sort the dataset based on the "Remarks" column
remarks_1 = dataset[dataset['Remarks'] == 1]
remarks_0 = dataset[dataset['Remarks'] == 0]

# Ensure equal number of rows for both Remarks
min_rows = min(len(remarks_1), len(remarks_0))
remarks_1 = remarks_1.head(min_rows)
remarks_0 = remarks_0.head(min_rows)

# Create an empty DataFrame to store the sorted rows
sorted_dataset = pd.DataFrame(columns=dataset.columns)

# Merge the DataFrames alternatively
for index in range(min_rows):
    sorted_dataset = sorted_dataset.append(remarks_1.iloc[index], ignore_index=True)
    sorted_dataset = sorted_dataset.append(remarks_0.iloc[index], ignore_index=True)

# Save the sorted dataset to a new CSV file
output_file_path = r"C:\Users\ASUS\Downloads\Final_train.csv"
sorted_dataset.to_csv(output_file_path, index=False)

print("Sorted dataset saved successfully.")
```
